```python
def create_data_list(dframe, col_ref):
    data_list = []
    for i in range(len(dframe)):
        brgy = dframe['brgy'].iloc[i]
        val = dframe[col_ref].iloc[i]
        if int(val) > 0:
            #remove if statement when all latloc is complete
            if brgy<120:
                #print(brgy, val, df_brgy[df_brgy['brgy']==brgy]['lat'].values, df_brgy[df_brgy['brgy']==brgy]['long'].values)
                #temp.append(val)
                #temp.append(df_brgy[df_brgy['brgy']==brgy]['lat'].values)
                #temp.append(df_brgy[df_brgy['brgy']==brgy]['long'].values)
                data_list.append([df_brgy[df_brgy['brgy']==brgy]['lat'].values[0], df_brgy[df_brgy['brgy']==brgy]['long'].values[0],float(val)])
    return data_list

#example usage
ndata = create_data_list(df_pnp, 'CONSUMATED THEFT(PROPERTY)')
ndata
```
